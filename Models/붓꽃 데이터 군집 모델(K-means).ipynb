```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

iris = load_iris()
iris_df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
iris_df['label'] = iris.target

X = iris_df.iloc[:, :4]
```

## 3개 군집으로 나누는 모델

```python
kmeans_3 = KMeans(n_clusters = 3, random_state = 42)
y_preds = kmeans_3.fit_predict(X)

iris_df['y_preds(3 Group)'] = y_preds
iris_df.head()
```

```python
iris_df.groupby(['label', 'y_preds(3 Group)'])['sepal length (cm)'].count()
```

## 2개 군집으로 나누는 모델

```python
kmeans_2 = KMeans(n_clusters = 2, random_state = 42)
y_preds = kmeans_2.fit_predict(X)

iris_df['y_preds(2 Group)'] = y_preds
iris_df.head()
```

```python
iris_df.groupby(['label', 'y_preds(2 Group)'])['sepal length (cm)'].count()
```
